```python
X = np.array([0,0.22,0.24,0.33,0.37,0.44,0.44,0.57,0.93,1])
y = np.array([0,0.22,0.58,0.20,0.55,0.39,0.54,0.53,1,0.61])
```

```python
import numpy as np

def gradient_descent(x,y):
    m_curr = 0.75
    b_curr = 0.45
    iterations = 100
    n = len(x)
    learning_rate = 0.01
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/2) * sum([val**2 for val in (y-y_predicted)])
        SSE = 1/(2*(sum((y-y_predicted)*(y-y_predicted))))
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print ("m {}, b {}, cost {},SSE {} iteration {}".format(m_curr,b_curr,cost,SSE, i))
```

```python
gradient_descent(X,y)
```

```python
def gradientDescent2(X, y, theta, alpha, num_iters):
    m = size(y)
    n = size(theta)
    tmp = zeros((n,1))
    for iter in xrange(1, num_iters + 1):
        for initer in xrange(0, n):
            tmp[initer] = theta[initer] - alpha * (1.0 / m) * sum(transpose(X * theta - y) * X[:,initer])
        theta = tmp
    return theta        
```
